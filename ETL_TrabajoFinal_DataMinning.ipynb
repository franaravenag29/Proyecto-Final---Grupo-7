```python
# ============================================================
# BLOQUE 1: Preparación del entorno de trabajo
# Proyecto Final Data Mining - CASEN 2024
# ============================================================

# Librerías principales
import pandas as pd
import numpy as np
import os
import random

# Configuración general
SEED = 123
np.random.seed(SEED)
random.seed(SEED)

# Mostrar más columnas y filas en las salidas
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", lambda x: f"{x:,.3f}")

print("Librerías cargadas correctamente.")
print(f"Semilla fijada: {SEED}")
```

```python
# ============================================================
# BLOQUE 2: Conexión con Google Drive y creación de carpetas
# ============================================================

from google.colab import drive
drive.mount('/content/drive')

# Ruta principal del proyecto en Google Drive
# Puedes cambiar el nombre de la carpeta si tu grupo quiere otro.
ruta_proyecto = "/content/drive/MyDrive/DataMining_CASEN2024"

# Carpetas internas del proyecto
ruta_datos_originales = os.path.join(ruta_proyecto, "datos_originales")
ruta_datos_procesados = os.path.join(ruta_proyecto, "datos_procesados")
ruta_documentos = os.path.join(ruta_proyecto, "documentos")
ruta_salidas = os.path.join(ruta_proyecto, "salidas")

# Crear carpetas si no existen
carpetas = [
    ruta_proyecto,
    ruta_datos_originales,
    ruta_datos_procesados,
    ruta_documentos,
    ruta_salidas
]

for carpeta in carpetas:
    os.makedirs(carpeta, exist_ok=True)

print("Google Drive conectado correctamente.")
print("Carpetas del proyecto creadas o verificadas:")
for carpeta in carpetas:
    print("-", carpeta)
```

```python
# ============================================================
# BLOQUE 3: Verificación de archivos en datos_originales
# ============================================================

# Listar archivos disponibles en la carpeta de datos originales
archivos_originales = os.listdir(ruta_datos_originales)

print("Archivos encontrados en datos_originales:")
for archivo in archivos_originales:
    print("-", archivo)

# Verificar específicamente si está la base CASEN 2024
nombre_base = "casen_2024.dta"
ruta_base_original = os.path.join(ruta_datos_originales, nombre_base)

if os.path.exists(ruta_base_original):
    print("\nBase encontrada correctamente:")
    print(ruta_base_original)
else:
    print("\nNo se encontró la base casen_2024.dta.")
    print("Revisa que el archivo esté en la carpeta datos_originales y que el nombre sea exactamente casen_2024.dta")
```

```python
# ============================================================
# BLOQUE 4: Revisión inicial de la estructura de la base CASEN
# ============================================================

# Abrir la base como lector, sin cargarla completa
lector_stata = pd.read_stata(
    ruta_base_original,
    iterator=True,
    convert_categoricals=False
)

# Leer solo las primeras 5 filas
df_muestra = lector_stata.get_chunk(5)

# Obtener nombres de variables desde la muestra
variables_base = df_muestra.columns.tolist()

# Intentar obtener etiquetas/descripciones de variables
try:
    etiquetas_variables = lector_stata.variable_labels()
except:
    etiquetas_variables = {}

print("Cantidad de variables detectadas:", len(variables_base))

print("\nPrimeras 50 variables encontradas:")
for var in variables_base[:50]:
    etiqueta = etiquetas_variables.get(var, "Sin etiqueta")
    print(f"- {var}: {etiqueta}")

print("\nDimensión de la muestra cargada:", df_muestra.shape)

display(df_muestra.head())
```

```python
# ============================================================
# BLOQUE 5: Verificación de variables seleccionadas para el ETL
# ============================================================

# Variables seleccionadas inicialmente según literatura y enfoque del proyecto
variables_etl = [
    # Identificación y control
    "folio",
    "id_persona",
    "expr",
    "pco1_a",

    # Variable objetivo
    "pobreza",

    # Demografía de la jefatura
    "edad",
    "sexo",
    "ecivil",

    # Educación
    "esc",

    # Trabajo
    "activ",
    "contrato",
    "cotiza",

    # Vulnerabilidad / identidad / migración
    "disc_wg",
    "pueblos_indigenas",
    "lugar_nac",

    # Composición del hogar
    "numper",
    "tipohogar",
    "men18c",
    "may60c",

    # Situación laboral del hogar
    "n_ocupados",
    "n_desocupados",
    "n_inactivos",

    # Territorio
    "region",
    "area",

    # Vivienda
    "ind_hacina",
    "ind_san",
    "ind_mat",
    "ind_estado",
    "ind_cal_glob",
    "ten_viv_f"
]

# Verificar cuáles variables están y cuáles no están en la base
variables_presentes = [var for var in variables_etl if var in variables_base]
variables_faltantes = [var for var in variables_etl if var not in variables_base]

print("Cantidad de variables propuestas:", len(variables_etl))
print("Cantidad de variables encontradas:", len(variables_presentes))
print("Cantidad de variables faltantes:", len(variables_faltantes))

print("\nVariables encontradas:")
for var in variables_presentes:
    print("-", var)

print("\nVariables faltantes:")
if len(variables_faltantes) == 0:
    print("No hay variables faltantes. Todas las variables propuestas existen en la base.")
else:
    for var in variables_faltantes:
        print("-", var)
```

```python
# ============================================================
# BLOQUE 6: Carga de variables seleccionadas desde CASEN 2024
# ============================================================

# Cargar solo las variables seleccionadas para el proyecto
df = pd.read_stata(
    ruta_base_original,
    columns=variables_etl,
    convert_categoricals=False
)

print("Base reducida cargada correctamente.")
print("Dimensión de la base reducida:")
print(df.shape)

print("\nPrimeras filas de la base reducida:")
display(df.head())

print("\nTipos de datos:")
display(df.dtypes)
```

```python
# ============================================================
# BLOQUE 7: Filtrar jefatura de hogar
# ============================================================

# Revisar distribución de la variable pco1_a
print("Distribución de pco1_a antes del filtro:")
display(df["pco1_a"].value_counts(dropna=False).sort_index())

# Filtrar solo personas que son jefatura de hogar
# Según la estructura observada, pco1_a = 1 identifica a la jefatura del hogar
df_hogares = df[df["pco1_a"] == 1].copy()

print("\nBase filtrada a nivel de hogares, usando jefatura de hogar.")
print("Dimensión antes del filtro:", df.shape)
print("Dimensión después del filtro:", df_hogares.shape)

# Verificar si queda un solo registro por hogar
duplicados_folio = df_hogares["folio"].duplicated().sum()

print("\nCantidad de folios duplicados después del filtro:", duplicados_folio)

print("\nPrimeras filas de la base filtrada:")
display(df_hogares.head())
```

```python
# ============================================================
# BLOQUE 8: Creación de la variable objetivo pobreza_ingresos
# ============================================================

# Revisar distribución original de la variable pobreza
print("Distribución original de la variable pobreza:")
display(df_hogares["pobreza"].value_counts(dropna=False).sort_index())

print("\nProporción original de la variable pobreza:")
display((df_hogares["pobreza"].value_counts(dropna=False, normalize=True).sort_index() * 100).round(2))

# Crear variable objetivo binaria
# pobreza = 1: pobreza extrema
# pobreza = 2: pobreza no extrema
# pobreza = 3: fuera de pobreza
#
# Nueva variable:
# pobreza_ingresos = 1 si el hogar está en pobreza extrema o pobreza no extrema
# pobreza_ingresos = 0 si el hogar está fuera de pobreza

df_hogares["pobreza_ingresos"] = np.where(
    df_hogares["pobreza"].isin([1, 2]), 1,
    np.where(df_hogares["pobreza"] == 3, 0, np.nan)
)

# Revisar distribución de la nueva variable objetivo
print("\nDistribución de la variable objetivo pobreza_ingresos:")
display(df_hogares["pobreza_ingresos"].value_counts(dropna=False).sort_index())

print("\nProporción de la variable objetivo pobreza_ingresos:")
display((df_hogares["pobreza_ingresos"].value_counts(dropna=False, normalize=True).sort_index() * 100).round(2))

# Verificar cantidad de valores perdidos en la variable objetivo
nulos_objetivo = df_hogares["pobreza_ingresos"].isna().sum()

print("\nValores perdidos en pobreza_ingresos:", nulos_objetivo)

# Interpretación automática simple
prop_pobreza = df_hogares["pobreza_ingresos"].mean() * 100

print(f"\nPorcentaje de hogares clasificados en pobreza por ingresos: {prop_pobreza:.2f}%")

if prop_pobreza < 20:
    print("Advertencia: la clase pobreza parece ser minoritaria. No se debe evaluar el modelo solo con accuracy.")
else:
    print("La clase pobreza no parece extremadamente minoritaria, pero igualmente se usarán métricas complementarias.")
```

```python
# ============================================================
# BLOQUE 9: Revisión de valores perdidos por variable
# ============================================================

# Calcular cantidad y porcentaje de valores perdidos
resumen_nulos = pd.DataFrame({
    "variable": df_hogares.columns,
    "n_nulos": df_hogares.isna().sum().values,
    "porcentaje_nulos": (df_hogares.isna().mean().values * 100).round(2)
})

# Ordenar de mayor a menor porcentaje de nulos
resumen_nulos = resumen_nulos.sort_values(
    by="porcentaje_nulos",
    ascending=False
).reset_index(drop=True)

print("Resumen de valores perdidos por variable:")
display(resumen_nulos)

# Separar variables según nivel de valores perdidos
variables_sin_nulos = resumen_nulos[resumen_nulos["porcentaje_nulos"] == 0]
variables_pocos_nulos = resumen_nulos[
    (resumen_nulos["porcentaje_nulos"] > 0) &
    (resumen_nulos["porcentaje_nulos"] <= 10)
]
variables_muchos_nulos = resumen_nulos[resumen_nulos["porcentaje_nulos"] > 10]

print("\nVariables sin valores perdidos:", variables_sin_nulos.shape[0])
display(variables_sin_nulos)

print("\nVariables con pocos valores perdidos, entre 0% y 10%:", variables_pocos_nulos.shape[0])
display(variables_pocos_nulos)

print("\nVariables con más de 10% de valores perdidos:", variables_muchos_nulos.shape[0])
display(variables_muchos_nulos)
```

```python
# ============================================================
# BLOQUE 9.1: Revisión de etiquetas y categorías de variables con nulos
# ============================================================

# Abrir nuevamente la base como lector de Stata para recuperar metadatos
lector_stata = pd.read_stata(
    ruta_base_original,
    iterator=True,
    convert_categoricals=False
)

# Obtener etiquetas de variables
try:
    etiquetas_variables = lector_stata.variable_labels()
except:
    etiquetas_variables = {}

# Obtener etiquetas de valores/categorías
try:
    etiquetas_valores = lector_stata.value_labels()
except:
    etiquetas_valores = {}

# Variables que requieren revisión especial
variables_revision = ["contrato", "esc", "activ", "cotiza"]

print("Revisión de etiquetas de variables:")
for var in variables_revision:
    print("\n" + "="*60)
    print("Variable:", var)
    print("Etiqueta:", etiquetas_variables.get(var, "Sin etiqueta encontrada"))

    print("\nDistribución de valores en la base filtrada:")
    display(df_hogares[var].value_counts(dropna=False).sort_index())

print("\nDiccionarios de etiquetas de valores disponibles en la base:")
print(list(etiquetas_valores.keys())[:30])
```

```python
# ============================================================
# BLOQUE 9.2: Revisión de etiquetas de valores y patrón de nulos
# ============================================================

# Mostrar etiquetas de valores para variables relevantes
variables_con_etiquetas = [
    "pobreza",
    "pco1_a",
    "sexo",
    "ecivil",
    "activ",
    "contrato",
    "cotiza",
    "disc_wg",
    "pueblos_indigenas",
    "lugar_nac",
    "region",
    "area",
    "ind_hacina",
    "ind_san",
    "ind_mat",
    "ind_estado",
    "ind_cal_glob",
    "ten_viv_f"
]

print("Etiquetas de categorías disponibles:")
for var in variables_con_etiquetas:
    print("\n" + "="*70)
    print("Variable:", var)

    if var in etiquetas_valores:
        print(etiquetas_valores[var])
    else:
        print("No se encontró etiqueta de valores para esta variable.")

# Revisar relación entre condición de actividad y contrato
print("\n" + "="*70)
print("Cruce entre activ y contrato:")
tabla_activ_contrato = pd.crosstab(
    df_hogares["activ"],
    df_hogares["contrato"],
    dropna=False
)
display(tabla_activ_contrato)

# Revisar relación entre condición de actividad y cotización
print("\n" + "="*70)
print("Cruce entre activ y cotiza:")
tabla_activ_cotiza = pd.crosstab(
    df_hogares["activ"],
    df_hogares["cotiza"],
    dropna=False
)
display(tabla_activ_cotiza)

# Revisar cuántos nulos de contrato hay según condición de actividad
print("\n" + "="*70)
print("Nulos de contrato según condición de actividad:")
nulos_contrato_por_activ = df_hogares.groupby("activ")["contrato"].apply(lambda x: x.isna().sum())
display(nulos_contrato_por_activ)

# Revisar nulos de esc según pobreza_ingresos
print("\n" + "="*70)
print("Nulos de esc según pobreza_ingresos:")
nulos_esc_por_pobreza = df_hogares.groupby("pobreza_ingresos")["esc"].apply(lambda x: x.isna().sum())
display(nulos_esc_por_pobreza)
```

```python
# ============================================================
# BLOQUE 10: Tratamiento de valores faltantes y categorías especiales
# ============================================================

# Crear una copia de trabajo para limpieza
df_limpia = df_hogares.copy()

# ------------------------------------------------------------
# 1. Tratamiento de contrato
# ------------------------------------------------------------
# contrato:
# 1 = Sí
# 0 = No
# -88 = No sabe
# NaN = No aplica / sin información aplicable según flujo de encuesta
#
# Decisión: recodificar NaN como 2 = No aplica.
# No se eliminan estos registros porque representan una parte importante
# de la base y están asociados al flujo de preguntas laborales.

df_limpia["contrato"] = df_limpia["contrato"].fillna(2)

# ------------------------------------------------------------
# 2. Tratamiento de esc
# ------------------------------------------------------------
# esc: años de escolaridad.
# Tiene pocos valores perdidos, por lo que se imputa con la mediana.
# Esta decisión evita perder hogares y mantiene una imputación simple
# y reproducible para una variable numérica.

mediana_esc = df_limpia["esc"].median()
df_limpia["esc"] = df_limpia["esc"].fillna(mediana_esc)

# ------------------------------------------------------------
# 3. Verificación posterior al tratamiento
# ------------------------------------------------------------

print("Tratamiento de valores faltantes realizado.")
print(f"Mediana utilizada para imputar esc: {mediana_esc}")

print("\nValores perdidos después del tratamiento:")
display(df_limpia.isna().sum().sort_values(ascending=False))

print("\nDistribución de contrato después de recodificar NaN como 2 = No aplica:")
display(df_limpia["contrato"].value_counts(dropna=False).sort_index())

print("\nDistribución de esc después de imputación:")
display(df_limpia["esc"].describe())
```

```python
# ============================================================
# BLOQUE 11: Revisión de códigos especiales
# ============================================================

# En CASEN algunas variables usan códigos especiales como:
# -88 = No sabe
# 9   = No sabe / ns-nr en algunas variables
# 99  = No sabe / ns-nr en algunas variables
#
# Estos valores no aparecen como NaN, pero deben revisarse antes del modelamiento.

variables_codigos_especiales = [
    "contrato",
    "cotiza",
    "disc_wg",
    "pueblos_indigenas",
    "lugar_nac",
    "ind_hacina",
    "ind_san",
    "ind_mat",
    "ind_estado",
    "ind_cal_glob",
    "ten_viv_f"
]

codigos_especiales = [-88, 9, 99]

resumen_codigos_especiales = []

for var in variables_codigos_especiales:
    for codigo in codigos_especiales:
        n = (df_limpia[var] == codigo).sum()
        porcentaje = (n / len(df_limpia)) * 100

        if n > 0:
            resumen_codigos_especiales.append({
                "variable": var,
                "codigo": codigo,
                "n_casos": n,
                "porcentaje": round(porcentaje, 3)
            })

resumen_codigos_especiales = pd.DataFrame(resumen_codigos_especiales)

print("Resumen de códigos especiales encontrados:")
if resumen_codigos_especiales.empty:
    print("No se encontraron códigos especiales en las variables revisadas.")
else:
    display(resumen_codigos_especiales)

print("\nDistribución completa de variables con posibles códigos especiales:")
for var in variables_codigos_especiales:
    print("\n" + "="*60)
    print("Variable:", var)
    display(df_limpia[var].value_counts(dropna=False).sort_index())
```

```python
# ============================================================
# BLOQUE 12: Recodificación de códigos especiales
# ============================================================

# Copia de seguridad antes de recodificar códigos especiales
df_limpia_recod = df_limpia.copy()

# ------------------------------------------------------------
# Recodificación de códigos especiales "No sabe"
# ------------------------------------------------------------
# En CASEN algunas variables usan -88 para "No sabe".
# Para evitar que modelos interpreten -88 como valor numérico extremo,
# se recodifica como una categoría positiva dentro de cada variable.

# contrato:
# 0 = No
# 1 = Sí
# 2 = No aplica
# 3 = No sabe
df_limpia_recod["contrato"] = df_limpia_recod["contrato"].replace(-88, 3)

# cotiza:
# 0 = No
# 1 = Sí
# 2 = No sabe
df_limpia_recod["cotiza"] = df_limpia_recod["cotiza"].replace(-88, 2)

# disc_wg:
# 0 = Sin discapacidad
# 1 = Con discapacidad
# 2 = No sabe
df_limpia_recod["disc_wg"] = df_limpia_recod["disc_wg"].replace(-88, 2)

# lugar_nac:
# 0 = Nacido(a) en Chile
# 1 = Nacido(a) fuera de Chile
# 2 = No sabe
df_limpia_recod["lugar_nac"] = df_limpia_recod["lugar_nac"].replace(-88, 2)

# ind_hacina:
# 1 = Sin hacinamiento
# 2 = Hacinamiento medio
# 3 = Hacinamiento alto
# 4 = Hacinamiento crítico
# 5 = No sabe
df_limpia_recod["ind_hacina"] = df_limpia_recod["ind_hacina"].replace(-88, 5)

# Verificación posterior
variables_recodificadas = ["contrato", "cotiza", "disc_wg", "lugar_nac", "ind_hacina"]

print("Recodificación de códigos especiales realizada.")

for var in variables_recodificadas:
    print("\n" + "="*60)
    print("Distribución de", var, "después de recodificar:")
    display(df_limpia_recod[var].value_counts(dropna=False).sort_index())

# Verificar que ya no existan valores -88 en la base
n_menos_88 = (df_limpia_recod == -88).sum().sum()

print("\nCantidad total de valores -88 restantes en la base:", n_menos_88)
```

```python
# ============================================================
# BLOQUE 13: Creación de variables nuevas justificadas
# ============================================================

# Crear una nueva copia de trabajo
df_final = df_limpia_recod.copy()

# ------------------------------------------------------------
# 1. Jefatura mujer
# ------------------------------------------------------------
# sexo:
# 1 = Hombre
# 2 = Mujer
#
# Nueva variable:
# jefatura_mujer = 1 si la jefatura es mujer
# jefatura_mujer = 0 si la jefatura es hombre

df_final["jefatura_mujer"] = np.where(df_final["sexo"] == 2, 1, 0)

# ------------------------------------------------------------
# 2. Nacido fuera de Chile
# ------------------------------------------------------------
# lugar_nac:
# 0 = Nacido(a) en Chile
# 1 = Nacido(a) fuera de Chile
# 2 = No sabe
#
# Nueva variable:
# nacido_fuera_chile = 1 si nació fuera de Chile
# nacido_fuera_chile = 0 si nació en Chile
# nacido_fuera_chile = 2 si no sabe

df_final["nacido_fuera_chile"] = df_final["lugar_nac"]

# ------------------------------------------------------------
# 3. Tasas laborales del hogar
# ------------------------------------------------------------
# Estas variables permiten comparar hogares de distinto tamaño.
# Por ejemplo, no es lo mismo tener 2 ocupados en un hogar de 2 personas
# que 2 ocupados en un hogar de 6 personas.

df_final["tasa_ocupados_hogar"] = df_final["n_ocupados"] / df_final["numper"]
df_final["tasa_desocupados_hogar"] = df_final["n_desocupados"] / df_final["numper"]
df_final["tasa_inactivos_hogar"] = df_final["n_inactivos"] / df_final["numper"]

# ------------------------------------------------------------
# 4. Verificación de variables creadas
# ------------------------------------------------------------

variables_creadas = [
    "jefatura_mujer",
    "nacido_fuera_chile",
    "tasa_ocupados_hogar",
    "tasa_desocupados_hogar",
    "tasa_inactivos_hogar"
]

print("Variables nuevas creadas correctamente:")
print(variables_creadas)

print("\nDistribución de jefatura_mujer:")
display(df_final["jefatura_mujer"].value_counts(dropna=False).sort_index())

print("\nDistribución de nacido_fuera_chile:")
display(df_final["nacido_fuera_chile"].value_counts(dropna=False).sort_index())

print("\nResumen de tasas laborales creadas:")
display(df_final[
    ["tasa_ocupados_hogar", "tasa_desocupados_hogar", "tasa_inactivos_hogar"]
].describe())

print("\nDimensión actual de la base:")
print(df_final.shape)
```

```python
# ============================================================
# BLOQUE 14: Selección de variables finales para la base limpia
# ============================================================

# Variables finales que se conservarán en la base limpia
variables_finales = [
    # Identificación y apoyo descriptivo
    "folio",
    "expr",

    # Variable original de pobreza y variable objetivo
    "pobreza",
    "pobreza_ingresos",

    # Características de la jefatura del hogar
    "edad",
    "jefatura_mujer",
    "ecivil",
    "esc",
    "activ",
    "contrato",
    "cotiza",
    "disc_wg",
    "pueblos_indigenas",
    "nacido_fuera_chile",

    # Composición del hogar
    "numper",
    "tipohogar",
    "men18c",
    "may60c",

    # Situación laboral del hogar
    "n_ocupados",
    "n_desocupados",
    "n_inactivos",
    "tasa_ocupados_hogar",
    "tasa_desocupados_hogar",
    "tasa_inactivos_hogar",

    # Territorio
    "region",
    "area",

    # Vivienda
    "ind_hacina",
    "ind_san",
    "ind_mat",
    "ind_estado",
    "ind_cal_glob",
    "ten_viv_f"
]

# Crear base final
df_modelo = df_final[variables_finales].copy()

print("Base final creada correctamente.")
print("Dimensión de la base final:")
print(df_modelo.shape)

print("\nPrimeras filas de la base final:")
display(df_modelo.head())

print("\nValores perdidos en la base final:")
display(df_modelo.isna().sum().sort_values(ascending=False))

print("\nVerificación de variable objetivo:")
display(df_modelo["pobreza_ingresos"].value_counts(normalize=True).mul(100).round(2))
```

```python
# ============================================================
# BLOQUE 15: Separación entre base documentada y base de modelamiento
# ============================================================

# ------------------------------------------------------------
# 1. Base limpia documentada
# ------------------------------------------------------------
# Esta base conserva variables de identificación y respaldo.
# Sirve para trazabilidad, EDA y documentación del ETL.

df_limpia_documentada = df_modelo.copy()

# ------------------------------------------------------------
# 2. Base para modelamiento
# ------------------------------------------------------------
# Se eliminan variables que NO deben entrar como predictoras:
# - folio: identificador del hogar
# - expr: factor de expansión, útil para EDA pero no predictor principal
# - pobreza: variable original usada para crear pobreza_ingresos
#
# La variable pobreza_ingresos sí se conserva porque es la variable objetivo.

variables_excluir_modelo = ["folio", "expr", "pobreza"]

df_modelamiento = df_limpia_documentada.drop(columns=variables_excluir_modelo)

print("Base limpia documentada:")
print(df_limpia_documentada.shape)

print("\nBase para modelamiento:")
print(df_modelamiento.shape)

print("\nVariables excluidas del modelamiento:")
print(variables_excluir_modelo)

print("\nColumnas de la base para modelamiento:")
for col in df_modelamiento.columns:
    print("-", col)

print("\nVerificación de valores perdidos en base de modelamiento:")
display(df_modelamiento.isna().sum().sort_values(ascending=False))

print("\nDistribución de la variable objetivo en base de modelamiento:")
display(df_modelamiento["pobreza_ingresos"].value_counts(normalize=True).mul(100).round(2))
```

```python
# ============================================================
# BLOQUE 16: Validación final del ETL
# ============================================================

print("VALIDACIÓN FINAL DEL ETL")
print("="*70)

# ------------------------------------------------------------
# 1. Dimensiones finales
# ------------------------------------------------------------
print("\n1. Dimensiones finales")
print("Base limpia documentada:", df_limpia_documentada.shape)
print("Base para modelamiento:", df_modelamiento.shape)

# ------------------------------------------------------------
# 2. Verificación de duplicados
# ------------------------------------------------------------
print("\n2. Verificación de duplicados")

duplicados_documentada = df_limpia_documentada["folio"].duplicated().sum()

print("Folios duplicados en base documentada:", duplicados_documentada)

duplicados_modelamiento = df_modelamiento.duplicated().sum()
print("Filas completamente duplicadas en base de modelamiento:", duplicados_modelamiento)

# ------------------------------------------------------------
# 3. Verificación de valores perdidos
# ------------------------------------------------------------
print("\n3. Verificación de valores perdidos")

nulos_documentada = df_limpia_documentada.isna().sum().sum()
nulos_modelamiento = df_modelamiento.isna().sum().sum()

print("Total de valores perdidos en base documentada:", nulos_documentada)
print("Total de valores perdidos en base de modelamiento:", nulos_modelamiento)

# ------------------------------------------------------------
# 4. Verificación de la variable objetivo
# ------------------------------------------------------------
print("\n4. Distribución de la variable objetivo")

tabla_objetivo = pd.DataFrame({
    "frecuencia": df_modelamiento["pobreza_ingresos"].value_counts().sort_index(),
    "porcentaje": df_modelamiento["pobreza_ingresos"].value_counts(normalize=True).sort_index().mul(100).round(2)
})

display(tabla_objetivo)

# ------------------------------------------------------------
# 5. Revisión de variables numéricas
# ------------------------------------------------------------
print("\n5. Resumen de variables numéricas")

variables_numericas_revision = [
    "edad",
    "esc",
    "numper",
    "n_ocupados",
    "n_desocupados",
    "n_inactivos",
    "tasa_ocupados_hogar",
    "tasa_desocupados_hogar",
    "tasa_inactivos_hogar"
]

display(df_modelamiento[variables_numericas_revision].describe())

# ------------------------------------------------------------
# 6. Verificación de rangos de tasas laborales
# ------------------------------------------------------------
print("\n6. Verificación de rangos de tasas laborales")

tasas = [
    "tasa_ocupados_hogar",
    "tasa_desocupados_hogar",
    "tasa_inactivos_hogar"
]

for var in tasas:
    minimo = df_modelamiento[var].min()
    maximo = df_modelamiento[var].max()
    fuera_rango = ((df_modelamiento[var] < 0) | (df_modelamiento[var] > 1)).sum()

    print(f"{var}: mínimo = {minimo:.3f}, máximo = {maximo:.3f}, valores fuera de [0,1] = {fuera_rango}")

# ------------------------------------------------------------
# 7. Distribución de variables categóricas
# ------------------------------------------------------------
print("\n7. Distribución de variables categóricas principales")

variables_categoricas_revision = [
    "jefatura_mujer",
    "ecivil",
    "activ",
    "contrato",
    "cotiza",
    "disc_wg",
    "pueblos_indigenas",
    "nacido_fuera_chile",
    "tipohogar",
    "region",
    "area",
    "ind_hacina",
    "ind_san",
    "ind_mat",
    "ind_estado",
    "ind_cal_glob",
    "ten_viv_f"
]

for var in variables_categoricas_revision:
    print("\n" + "="*60)
    print("Variable:", var)
    display(df_modelamiento[var].value_counts(dropna=False).sort_index())

print("\nValidación final terminada.")
```
